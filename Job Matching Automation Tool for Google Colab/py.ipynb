```python
import json

# Load profile data
with open("profile_data.json", "r") as file:
    profile = json.load(file)

roles = profile['Desired Roles']
cities = profile['Desired City']

# Define base job search URL (this is a placeholder, use real job board API endpoints)
#add linkdin, indeed, glassdoor, etc.
job_search_base_url = "https://www.linkedin.com/search/results/jobs/"
# Prepare queries
queries = []
for city in cities:
    for role in roles:
        query = {
            "role": role,
            "city": city
        }
        queries.append(query)

# Example function to perform a search (requires proper authentication and API usage)
#only run 100 turns

def search_jobs(query):
    # Simulate sending request (replace with real requests.get or similar)
    print(f"Searching for {query['role']} jobs in {query['city']}...")
    # Return dummy results for illustration
    return [{"title": f"{query['role']} at ExampleCorp", "location": query["city"]}]

# Limit to 100 queries only
queries = queries[:100]

# Run all job searches
all_results = []
for q in queries:
    results = search_jobs(q)
    all_results.extend(results)

print(all_results)

```

```python
# Example output
print("Job search completed. Found results:")
for result in all_results:
    print(f"- {result['title']} in {result['location']}")
```

```python
import csv

# Write all_results to a CSV file
with open('job_search_results.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['title', 'location']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(all_results)

print("All job search results have been saved to job_search_results.csv")
```
