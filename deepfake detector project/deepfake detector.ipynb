```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the saved model
model = tf.keras.models.load_model('D:\codes\my-projects\deepfake detector project\MODELdeepfakedetector.h5')

# Function to predict if an image is fake or real and return confidence rate
def predict_image(image_path):
    # Load and preprocess the image
    img = image.load_img(image_path, target_size=(128, 128))  # Resize to match model input size
    img = image.img_to_array(img)
    img = img / 255.0  # Normalize the image
    img = np.expand_dims(img, axis=0)  # Add batch dimension

    # Make prediction
    prediction = model.predict(img)[0][0]  # Get the probability of being "Real"
    confidence = float(prediction)  # Convert to Python float

    # Determine the label and confidence rate
    if confidence < 0.5:
        label = "Fake"
        confidence_rate = 1 - confidence  # Confidence for "Fake"
    else:
        label = "Real"
        confidence_rate = confidence  # Confidence for "Real"

    return label, confidence_rate
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os
from tkinter import Tk, filedialog

# Load the saved model
model = tf.keras.models.load_model('D:\codes\my-projects\deepfake detector project\MODELdeepfakedetector.h5')

# Function to predict if an image is fake or real and return confidence rate
def predict_image(image_path):
    # Load and preprocess the image
    img = image.load_img(image_path, target_size=(128, 128))  # Resize to match model input size
    img = image.img_to_array(img)
    img = img / 255.0  # Normalize the image
    img = np.expand_dims(img, axis=0)  # Add batch dimension

    # Make prediction
    prediction = model.predict(img)[0][0]  # Get the probability of being "Real"
    confidence = float(prediction)  # Convert to Python float

    # Determine the label and confidence rate
    if confidence < 0.5:
        label = "Fake"
        confidence_rate = 1 - confidence  # Confidence for "Fake"
    else:
        label = "Real"
        confidence_rate = confidence  # Confidence for "Real"

    return label, confidence_rate

# Function to upload an image and make a prediction
def upload_and_predict():
    print("Please select an image to check if it's real or fake...")
    root = Tk()
    root.withdraw()  # Hide the main window
    file_path = filedialog.askopenfilename(
        title="Select image",
        filetypes=[("Image files", "*.jpg *.jpeg *.png *.bmp *.gif")]
    )
    root.destroy()

    if not file_path:
        print("No file selected.")
        return

    filename = os.path.basename(file_path)
    print(f"Analyzing {filename}...")

    # Display the uploaded image
    plt.figure(figsize=(6, 6))
    img = plt.imread(file_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Uploaded Image: {filename}")
    plt.show()

    # Get the prediction and confidence rate
    label, confidence_rate = predict_image(file_path)

    # Display the result
    print(f"The image is predicted to be: {label} (Confidence: {confidence_rate:.2f})")

    # Add visual indication of the result
    color = 'green' if label == "Real" else 'red'

    plt.figure(figsize=(8, 2))
    plt.text(0.5, 0.5, f"PREDICTION: {label.upper()} (Confidence: {confidence_rate:.2f})",
             fontsize=16, ha='center', color=color, weight='bold')
    plt.axis('off')
    plt.show()

# Call the function to upload and predict
upload_and_predict()
```
